```python
# pip install google-search-results
from langchain_community.utilities import SerpAPIWrapper

params = {
    "engine":"google_news",
    "gl":"kr",  # Geo Location
    "hl":"ko"   # host Language
}
search = SerpAPIWrapper(params = params)
search.run("이차전지 산업")
    
```

```python
results = search.run("로봇 산업")
len(results)
```

```python
results[0]
# position, title, source, icon, link, thumbnail, date
```

```python
results[0].keys()
```

```python
url=results[0]['link']
url
```

```python
# 웹 페이지에서 데이터 가져오기
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(url)
docs = loader.load()
docs[0]
```

```python
# 뉴스기사에서 본문 텍스트 추출
def extract_longest_text(text):
    segments = text.split('\n')
    longest_segment = max(segments, key=len)
    return longest_segment

text = docs[0].page_content
longest_text = extract_longest_text(text)
print(longest_text)
```

```python
# SerpAPI 검색 결과를 순회하면서 본문 텍스트 추출
from tqdm import tqdm
import time

new_results = []
for result in tqdm(results):
    try:
        url = result['link']
        loader = WebBaseLoader(url)
        docs = loader.load()
        text = docs[0].page_content
        longest_text = extract_longest_text(text)
        result['content']=longest_text
        new_results.append(result)
    except:
        pass

print("new_results 개수: ", len(new_results))
new_results[0]
```

```python
import pandas as pd
data = pd.DataFrame(new_results)
data.head()
data.info()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

prompt_template = """ 
Write a concise summary of the fllowing in korean Hangul(한글) :
"{text}"
CONCISE SUMMARY : """

prompt = PromptTemplate.from_template(prompt_template)
llm = ChatOpenAI(temperature=0, model_name="gpt-4o-mini")
output_parser = StrOutputParser()
llm_chain = prompt | llm | output_parser

response = llm_chain.invoke({"text": data['content'][0]})
response
```

```python
# 뉴스 본문을 입력으로 사용해서 요약 결과를 생성하는 함수
def summarize_news(content):
    response = llm_chain.invoke({"text": content})
    return response

df_test = data.head(3)
df_test['summary'] = df_test['content'].apply(summarize_news)
df_test[['content','summary']]
```

```python
#뉴스 본문에서 3개의 키워드를 추출하는 LLM Chain 구현
from langchain.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

# prompt
prompt_template = """Please extract 3 key words from the following content in Korean Hangul (한글) and separate them with commas (,):
"{text}"
Key words:"""

prompt = PromptTemplate.from_template(prompt_template)

# LLM
llm = ChatOpenAI(temperature=0, model_name="gpt-3.5-turbo-0125")

# output parser
output_parser = StrOutputParser()

# Chain
llm_chain = prompt | llm | output_parser

response = llm_chain.invoke({"text": data['content'][0]})

response
```

```python
# 뉴스 본문을 입력으로 사용하여 핵심 키워드를 추출하는 함수
def extract_keywords(content):
    response = llm_chain.invoke({"text": content})
    return response

# 결과 확인 - 테스트를 위해서 첫 3행만 별도로 추출하여 추출
df_test['keywords'] = df_test['content'].apply(extract_keywords)

df_test[['content', 'summary', 'keywords']]
```
